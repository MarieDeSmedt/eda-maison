# Réalisez un code permettant d'identifier automatiquement les variables de type Object

```python
df.select_dtypes(include='object').count()
```

```python
# je cré un dataframe avec les colonnes de type object uniquement
s_object = df.select_dtypes(include='object')
columns = []
for title in s_object:
    columns.append(title)
#print(columns)
df_object = df.loc[:,columns]
print(df_object.shape)
```
